```python
def profit_calculation(Date_and_Choice,Date_Profit_Rate):
    for i in range(0,len(Date_and_Choice.index)):
        profit = 0
        if i > 0:
            Choices = Date_and_Choice[Date_and_Choice.index[i-1]]
            if len(Choices) > 0:
                for j in range(1,len(Choices)+1):
                    cut = len(Choices)*(1+len(Choices)) / 2
                    coe = len(Choices) - j + 1
                    if Choices[j-1] in dfs[''].columns:
                        profit = profit + (coe/cut)*dfs[''].loc[Date_and_Choice.index[i],Choices[j-1]]
                    else:
                        profit = profit + (coe/cut)*dfs[''].loc[Date_and_Choice.index[i],Choices[j-1]]
            else:
                profit = 0
        else:
            profit = 0
        Date_Profit_Rate[Date_and_Choice.index[i]] = profit
        
#
def accumulate_profit_calculation(Date_Profit_Rate,Date_Accumulate_Profit_Rate):
    for i in range(0,len(Date_Profit_Rate.index)):
        if i == 0:
            Date_Accumulate_Profit_Rate[Date_Profit_Rate.index[i]] = 1*(1+Date_Profit_Rate[Date_Profit_Rate.index[i]]/100)
        else:
            Date_Accumulate_Profit_Rate[Date_Profit_Rate.index[i]] = Date_Accumulate_Profit_Rate[Date_Profit_Rate.index[i-1]]*(1+Date_Profit_Rate[Date_Profit_Rate.index[i]]/100)
```

```python
yi_ji = pd.read_excel('.xlsx')
yi_ji.set_index('',inplace=True)

percent_30_times_40 = pd.DataFrame(columns=dfs[''].columns,index=dfs[''].index)
percent_30_times_40.fillna(0,inplace=True)
for date in dfs[''].index:
    in_choice = dfs[''].loc[date,:].sort_values(ascending=False)[0:30].index
    percent_30_times_40.loc[date,in_choice] += 1
def Sum(series):
    return sum(series)

stats_times = percent_30_times_40.apply(sum)

all_list = []
for first_industry in yi_ji.index:
    loc = yi_ji.index.get_loc(first_industry)
    if loc < len(yi_ji.index)-1:
        next_first_industry = yi_ji.index[loc+1]
        num = first_industry[2:6]
        next_num = next_first_industry[2:6]
        judge_list = []
        for n in range(int(num),int(next_num)):
            if str(80)+str(n)+'.SI' in percent_30_times_40.columns:
                judge_list.append(str(80)+str(n)+'.SI')
        part_of_second_industry = stats_times[judge_list].sort_values(ascending=False)[0:round(len(judge_list)/2)].index
        for cho in part_of_second_industry:
            all_list.append(cho)
display(len(all_list))
```
